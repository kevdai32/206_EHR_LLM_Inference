## Extraction for med_gemma

```python
import json
import ijson

# Step 1: Load labeled JSON and filter for imatinib_mentioned = True
filtered_notes = set()
with open("open_hermes_test_added_Prompt.json", "r") as f:
    for item in ijson.items(f, "item"):
        if item.get("imatinib_mentioned", False):
            filtered_notes.add(item["original_note"])

print(f"Found {len(filtered_notes)} notes with imatinib mentioned")

# Step 2: Load original notes and extract person_source_value
person_source_values = set()
with open("latest_notes.json", "r") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        note = json.loads(line)
        if note["note_text"] in filtered_notes:
            person_source_values.add(note["person_source_value"])

# Step 3: Save as plain text, one per line
with open("imatinib_person_source_values.txt", "w") as f:
    for value in person_source_values:
        f.write(f"{value}\n")

print(f"Saved {len(person_source_values)} unique person_source_values to imatinib_person_source_values.txt")
```
